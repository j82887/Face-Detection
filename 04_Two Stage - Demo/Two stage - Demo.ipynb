```python
import cv2
from keras.models import Model
import keras
import numpy as np
        
#載入模型
model = keras.models.load_model('modelunsharpgb_out.h5')

#框住人臉的矩形邊框顏色與文字顏色(RGB轉換表:https://convertingcolors.com/rgb-color-255_255_255.html)  
color_0 = (255, 255, 0) #黃色
color_1 = (0, 0, 255) #藍色
color_2 = (0, 255, 0) #綠色
color_3 = (255,0,255) #紫色
color_4 = (255, 0, 0) #紅色
color_5 = (255,255,255) #白色
color_6 = (255, 97, 0) #橘色
color_7 = (173, 255, 47) #綠黃色
color_8 = (0, 255, 255) #水色
    
#捕獲指定攝影機的即時影訊流
#cap = cv2.VideoCapture(0)
    
#捕獲指定影片
cap = cv2.VideoCapture("測試影片/videoplayback3.mp4")
    
#循環檢測辨識人臉
while True:
    ret, frame = cap.read()   #讀取畫面
        
    if ret is True:
            
        #影像灰階化，減少計算複雜度
        frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    else:
        continue
        
    #讀取OpenCV的人臉辨認器              
    cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
    cascade.load('haarcascades/haarcascade_frontalface_default.xml')

    #給定分類器的參數(縮放尺度，最小鄰近數量與影像最小範圍)
    faceRects = cascade.detectMultiScale(frame_gray, scaleFactor = 1.2, minNeighbors = 5, minSize = (30, 30))        
        
    #大於零代表現在有抓到人臉(當前畫面可能不只有一張)
    if len(faceRects) > 0: 

        #將抓到的人臉用for迴圈一一辨識
        for faceRect in faceRects: 

            #此時的人臉的位置
            x, y, w, h = faceRect
                
            #擷取這位置的影像存在image的變數
            image = frame[y - 10: y + h + 10, x - 10: x + w + 10]

            #將image轉為灰階
            image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

            #將image轉為大小為32×32
            image = cv2.resize(image,(32,32),interpolation=cv2.INTER_CUBIC)
                
            #將image轉為np的陣列並且轉為浮點數，進行正規化後存在data
            data = (np.asarray(image, dtype="uint8").astype('float32'))/255

            #由於模型的預測是多張的，所以需要增加一維度，因此使用reshape改變形狀
            data = data.reshape((1, 32, 32, 1)) 
                
            #即時影像使用模型進行預測，其中predict_classes是回傳最高機率的Label，predict是回傳所有類別的機率
            #faceID = model.predict_classes(data)   
            faceID = model.predict(data) 

            #條件式如果符合該類別的機率高於閥值，則框出人臉並顯示文字 
            #第1個人的條件式               
            if faceID[0][0] > 0.9999:     

                #人臉框出                                   
                cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), color_0, thickness = 2)
                    
                #文字提示 (畫面, 放入的字, 座標, 字體, 字體大小的縮放, 顏色, 字的粗度)
                cv2.putText(frame,'Class 0: Qiu Feng Ze', (10, 15), cv2.FONT_HERSHEY_TRIPLEX, 0.5, color_0, 1)

            elif faceID[0][1] > 0.9999:                                                        
                cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), color_1, thickness = 2)
                cv2.putText(frame,'Class 1: Nine Chen', (10, 15), cv2.FONT_HERSHEY_TRIPLEX, 0.5, color_1, 1) 

            elif faceID[0][2] > 0.9999:                                                        
                cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), color_2, thickness = 2)
                cv2.putText(frame,'Class 2: David Chao', (10, 15), cv2.FONT_HERSHEY_TRIPLEX, 0.5, color_2, 1)
                
            elif faceID[0][3] > 0.9999:                                                        
                cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), color_3, thickness = 2)
                cv2.putText(frame,'class 3: Butterfly', (10, 15), cv2.FONT_HERSHEY_TRIPLEX, 0.5, color_3, 1) 
                
            elif faceID[0][4] > 0.9999:                                                        
                cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), color_4, thickness = 2)
                cv2.putText(frame,'Class 4: Esther Wu', (10, 15), cv2.FONT_HERSHEY_TRIPLEX, 0.5, color_4, 1) 
                
            elif faceID[0][5] > 0.9999:                                                        
                cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), color_5, thickness = 2)
                cv2.putText(frame,'Class 5: Lai', (10, 15), cv2.FONT_HERSHEY_TRIPLEX, 0.5, color_5, 1)   
                
            elif faceID[0][6] > 0.9999:                                                        
                cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), color_6, thickness = 2)
                cv2.putText(frame,'Class 6: Yako Chan', (10, 15), cv2.FONT_HERSHEY_TRIPLEX, 0.5, color_6, 1)  
                
            elif faceID[0][7] > 0.9999:                                                        
                cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), color_7, thickness = 2)
                cv2.putText(frame,'Class 7: Cathy Shyu', (10, 15), cv2.FONT_HERSHEY_TRIPLEX, 0.5, color_7, 1) 
                
            elif faceID[0][8] > 0.9999:                                                        
                cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), color_8, thickness = 2)
                cv2.putText(frame,'Class 8: JunShou Lou' , (10, 15), cv2.FONT_HERSHEY_TRIPLEX, 0.5, color_8, 1)  
                
            else:
                pass

    cv2.imshow("Demo", frame)
        
    #等待1毫秒看是否有按鍵輸入
    k = cv2.waitKey(1)
    #如果輸入q則退出循環
    if k & 0xFF == ord('q'):
        break

#釋放影像並銷毀所有窗口
cap.release()
cv2.destroyAllWindows()
```
